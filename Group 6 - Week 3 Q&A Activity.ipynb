```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = pd.read_csv("diabetes_model.csv")
d.head(20)
```

## <font color="#ffa008" >Count plot of Outcome</font>

```python
#Count plot on outcome
sns.countplot(x='Outcome',data = d)
plt.show()
```

## <font color="#ffa008" >Box plot Age vs Outcome</font>

```python
#Box plot age vs outcome
sns.boxplot(x = 'Outcome', y= 'Age',data = d)
plt.title("Age vs Outcome")
plt.show()
```

## <font color="#ffa008" >Check for the chances of getting diabetes if it is common in family, using the Daiabetes Pedigree Function</font>

```python
#Check for the chances of getting diabetes if it is common in family, using the Daiabetes Pedigree Function

sns.boxplot(x='Outcome', y= 'DiabetesPedigreeFunction',data=d)
plt.title("DPF vs Outcome")
plt.show()
```

```python

sns.swarmplot(x = 'Outcome', y="DiabetesPedigreeFunction",data = d)
```

# <font color="ff0066" >Check for the relationship with Insulin & Glucose and group the datapoints by Outcome</font>

```python
#Check for the relationship with Insulin & Glucose and group the datapoints by Outcome

plt.figure(figsize=(10,5))
sns.scatterplot(x=d['Insulin'],y=d['Glucose'],s=15,hue=d['Outcome'],data=d)
plt.title('Plot of Insulin vs Glucose',fontsize=14)
plt.xlabel('Insulin')
plt.ylabel('Glucose')
```

### 5. Create a heatmap on Correlation of features in the data.

```python
#Create a heatmap on Correlation of features in the data
plt.figure(figsize=(10,10))
sns.heatmap(d.corr(),annot = True)
```
